# Unite

Similar to 'Download_unite_interval' from the commercial cleaning, the purpose of this script is to join together the separate files in the bucket. 
In this case the 3 files cover three time periods: 2015-2018, 2019 Jan-May, and 2019 June.

Note this was run from Siobhan's SLAC laptop where "aws configure" is set up.

```python
import s3fs
import pandas as pd
import boto3
from io import StringIO
```

Files are saved with the date this script was run, since as new data comes in this will be updated

```python
save_date = '26072019'
```

# Sessions Data

```python
s0 = pd.read_csv('s3://script.chargepoint.data/Residential_Data/Charging_Session_SLAC_SingleFamilyResidential.csv000')
s1 = pd.read_csv('s3://script.chargepoint.data/Residential_Data/Charging_Session_SLAC_Home_201901to201905.csv000')
s2 = pd.read_csv('s3://script.chargepoint.data/Residential_Data/Charging_Session_SLAC_Home_201906.csv000')
```

```python
# check before this stage that all the column headers are the same
s = pd.concat([s0, s1, s2], axis=0, ignore_index=True)
s = s.sort_values(by='Station Start Time (Local)')
s = s.reset_index(drop=True)
s
```

```python
csv_buffer = StringIO()
s.to_csv(csv_buffer, index=False)
s3_resource = boto3.resource('s3')
s3_resource.Object('script.chargepoint.data', 'Residential_Data/clean/Charging_Sessions_United'+str(save_date)+'.csv').put(Body=csv_buffer.getvalue())
```

# Intervals Data

Can restart kernel here to free up space if needed (don't forget to re-run box with import statements).

```python
i0 = pd.read_csv('s3://script.chargepoint.data/Residential_Data/Charging_Session_Interval_SLAC_SingleFamilyResidential.csv000')
i1 = pd.read_csv('s3://script.chargepoint.data/Residential_Data/Charging_Session_Interval_SLAC_Home_201901to201905.csv000')
i2 = pd.read_csv('s3://script.chargepoint.data/Residential_Data/Charging_Session_Interval_SLAC_Home_201906.csv000')
```

```python
i_united = pd.concat([i0, i1, i2], axis=0, ignore_index=True)
i_united = i_united.reset_index(drop=True)
i_united
```

```python
csv_buffer = StringIO()
i_united.to_csv(csv_buffer, index=False)
s3_resource = boto3.resource('s3')
s3_resource.Object('script.chargepoint.data', 'Residential_Data/clean/Charging_Intervals_United'+str(save_date)+'.csv').put(Body=csv_buffer.getvalue())
```
