```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
test = False
agg_by_port = False
reduce_ports = False
agg_by_driver = True
reduce_drivers = True

# Loading data into pandas dataframe
path = "../../data/clean/"
file = "merged.csv"
cols = ['Driver ID', 'Port ID', 'Interval Start Time (Local)', 'Interval Energy']
```

```python
df = pd.read_csv(path + file, index_col=False, usecols=cols)
```

```python
def to_datetime(x):
    return datetime.strptime(x, "%Y-%m-%d %H:%M:%S.%f")
def to_year(x):
    return x.year
```

```python
df['datetime'] = df['Interval Start Time (Local)'].apply(to_datetime)
df['year'] = df['datetime'].apply(to_year)
df["Interval Energy"] = df["Interval Energy"].round(4)
```

```python
df_copy = df.copy(deep=True)
```

# Grouping by Port and aggregating energy by time

```python
if agg_by_port:
    df = df[~df['Port ID'].isnull()]
    df['Port ID'] = df['Port ID'].apply(lambda x: str(int(x)))
```

```python
if agg_by_port and reduce_ports:
    df = df[~df['Port ID'].isnull()]
    ports = df['Port ID'].value_counts()
    # print(ports)
    print(len(ports))
    print(np.percentile(ports, 10))
    print(np.percentile(ports, 20))
    print(np.percentile(ports, 50))
    print(np.percentile(ports, 90))
    ports = ports[ports >= 500]
    print(len(ports))
    ports = [str(int(x)) for x in ports.index.values]
    df = df[df['Port ID'].isin(ports]]
```

```python
if agg_by_port:
    years = [2015, 2016, 2017, 2018]
    np_dict = {}
    for y in years:
        print("starting ", y)
        np_dict[y] = {}
        start_of_year = datetime(y, 1, 1)
        steps_in_year = 96 * (365 if y != 2016 else 366)
        def step_in_year(x):
            return int((x - start_of_year).total_seconds() // 900)

        df_year = df[df['year'] == y]
        print("applying step in year")
        df_year['step'] = df_year['datetime'].apply(step_in_year)

        df_year = df_year[["Port ID", "step", "Interval Energy"]]
        df_year_grouped = df_year.groupby(['Port ID'])
        print("aggregating")
        df_year_agg = df_year_grouped.agg(list)
        print("creating empty numpy arrays")
        for p in df_year_agg.index.values:
            np_dict[y][p] = np.zeros(steps_in_year)

        num_rows = len(df_year_agg)
        i = 0
        for index, row in df_year_agg.iterrows(): 
            i += 1
            if i % 100 == 0:
                print(i, " / ", num_rows)
            np_dict[y][index][row['step']] = row['Interval Energy']
```

```python
if agg_by_port:
    df_dict = {}
    for y in years:
        df_dict[y] = pd.DataFrame.from_dict(np_dict[y])
```

```python
if agg_by_port:
    for y in years:
        add = "_reduced" if reduce_ports else ""
        df_dict[y].to_csv(path + "ports_{}{}.csv".format(y, add), index = False)
```
